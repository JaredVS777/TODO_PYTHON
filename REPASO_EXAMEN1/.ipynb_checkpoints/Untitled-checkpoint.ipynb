TERCERA PREGUNTA

```python
archivo = open('jhon.txt', 'r', encoding='ANSI') # Abrir archivo
mensaje = archivo.read().lower()  # Convertir a minúsculas
archivo.close()
```

```python
palabras = mensaje.split()  # Separar el texto en palabras
palabras_unicas = set(palabras)  # Obtener las palabras únicas utilizando un conjunto
```

```python
fre_palabras = {} # Diccionario vacío
```

```python
for palabra in palabras:
    palabra = palabra.lower().strip(".,[...]{}...()  +*-") # Convertir a minúsculas y eliminar signos de puntuación
    if palabra in fre_palabras:
        fre_palabras[palabra] += 1 # Si la palabra ya está en el diccionario, aumentar su contador para indicar que se encontró una ocurrencia más
    else:
        fre_palabras[palabra] = 1 # Si la palabra no está en el diccionario, agregarla con un contador inicial de 1 para indicar que se encontró una ocurrencia
```

```python
# Ordenar las palabras alfabéticamente mediante el método sorted
palabras_ordenadas = sorted(fre_palabras.items()) # El método items() devuelve una lista de tuplas (clave, valor)
```

```python
# Imprimir los resultados
print("Listado de palabras y frecuencias:")
listado="" # Variable para almacenar el listado de palabras y frecuencias
for palabra, frecuencia in palabras_ordenadas: # Recorrer la lista de tuplas
    print(f"[{palabra} - {fre_palabras[palabra]}]") # Formato de salida de cada palabra y su frecuencia
```
